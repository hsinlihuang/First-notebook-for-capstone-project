```python
df = pd.read_csv('toronto.csv')
df.head()
```

#### read the csv file via panda

```python
c = pd.read_csv('https://cocl.us/Geospatial_data')
c.columns = ['PostalCode', 'Latitude', 'Longitude']
c.head()
```

#### merging the data

```python
df = df.merge(c, on='PostalCode')
df.head()
```

#### Take some row for example

```python
column_names = ["PostalCode", "Borough", "Neighborhood", "Latitude", "Longitude"]
test_df = pd.DataFrame(columns=column_names)

test_list = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for postcode in test_list:
    test_df = test_df.append(df[df["PostalCode"]==postcode], ignore_index=True)
    
test_df
```
